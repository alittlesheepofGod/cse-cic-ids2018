# Exploring and Clean Data of the CIC-IDS-2018 Dataset
[official dataset link](https://www.unb.ca/cic/datasets/ids-2018.html)


Based on the following exploration; The __"Friday-02-03-2018_TrafficForML_CICFlowMeter.csv"__, 
__"Friday-16-02-2018_TrafficForML_CICFlowMeter.csv"__, __"Friday-02-23-2018_TrafficForML_CICFlowMeter.csv"__, 
__"Thursday-01-03-2018_TrafficForML_CICFlowMeter.csv"__, __"Thursday-02-15-2018_TrafficForML_CICFlowMeter.csv"__, 
__"Thursday-02-22-2018_TrafficForML_CICFlowMeter.csv"__, __"Tuesday-02-20-2018_TrafficForML_CICFlowMeter.csv"__, 
__"Wednesday-02-14-2018_TrafficForML_CICFlowMeter.csv"__, __"Wednesday-02-21-2018_TrafficForML_CICFlowMeter.csv"__, 
__"Wednesday-02-28-2018_TrafficForML_CICFlowMeter.csv"__ csv files were chosen with the following labels:
* __Benign__ (1,209,156 total instances)
* __Bot__ (286,191 total instances)
* __DoS attacks-SlowHTTPTest__ (139,890 total instances)
* __DoS attacks-Hulk__ (461,912 total instances)
* __Brute Force -Web__ (1,209,156 total instances)
* __Brute Force -XSS__ (286,191 total instances)
* __SQL Injection__ (139,890 total instances)
* __DDoS attacks-LOIC-HTTP__ (461,912 total instances)
* __DoS attacks-GoldenEye__ (1,209,156 total instances)
* __Infilteration__ (286,191 total instances)
* __DoS attacks-Slowloris__ (139,890 total instances)
* __FTP-BruteForce__ (461,912 total instances)
* __SSH-Bruteforce__ (1,209,156 total instances)
* __DDOS attack-LOIC-UDP__ (286,191 total instances)
* __DDOS attack-HOIC__ (139,890 total instances)
* Label (one erroneous row, would need to be removed)

__Reasoning__: The labels are better balanced than the rest of the files. For the purposes of a baseline experiment; it would be satisfactory.

The __Bot__, __DoS attacks-SlowHTTPTest__, __DoS attacks-Hulk__, __Brute Force -Web__, __Brute Force -XSS__, __SQL Injection__, __DDoS attacks-LOIC-HTTP__, __DoS attacks-GoldenEye__, __Infilteration__, __DoS attacks-Slowloris__, __FTP-BruteForce__, __SSH-Bruteforce__, __DDOS attack-LOIC-UDP__, __DDOS attack-HOIC__ labels would be combined into one __"Malicious"__ label for a total of 887,993 __"Malicious"__ instances

The baseline experiment would be a binary classification experiment between __236,421 "Malicious"__ instances vs __63,489 "Benign"__ instances.
